```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Assuming you have selected relevant features for clustering
features_for_clustering = df[['Occupation', 'Age','Income','Education','Settlement size']]
```

```python
# Choose the number of clusters (k)
k = 5
```

```python
# Instantiate the KMeans model
kmeans = KMeans(n_clusters=k, random_state=42)
```

```python
# Fit the model to your data
kmeans.fit(features_for_clustering)
```

```python
KMeans(n_clusters=5, random_state=42)
```

```python

# Add cluster labels to your DataFrame
df['cluster'] = kmeans.labels_
```

```python

plt.scatter(features_for_clustering['Age'], features_for_clustering['Income'], c=df['cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s=100, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Age ')
plt.ylabel('Income')
plt.legend()
plt.show()
```

```python
wcss = []

# Try k values from 1 to 10 (you can adjust this range)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_for_clustering)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```
