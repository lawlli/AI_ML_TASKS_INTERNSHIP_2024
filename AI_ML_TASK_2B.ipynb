```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
numerical_columns = data.select_dtypes(include=[np.number]).columns.tolist()

# Calculate the number of rows needed for subplots
num_rows = (len(numerical_columns) + 1) // 2

plt.figure(figsize=(16, num_rows * 5))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(num_rows, 2, i)
    sns.histplot(data[column], kde=True, color='skyblue')
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
numerical_columns = data.select_dtypes(include=[np.number]).columns.tolist()

plt.figure(figsize=(16, 10))
for i, col1 in enumerate(numerical_columns, 1):
    for j, col2 in enumerate(numerical_columns, 1):
        if i < j:  # Plot only the lower triangular part of the matrix
            plt.subplot(len(numerical_columns) - 1, len(numerical_columns) - 1, (i - 1) * (len(numerical_columns) - 1) + j - 1)
            sns.scatterplot(x=col1, y=col2, data=data, color='skyblue')
            plt.title(f'{col1} vs {col2}')

plt.tight_layout()
plt.show()
```
